# Nurses Procurement with execution time

```python
data_nursing = {year: 0 for year in range(2015, 2026)}
```

```python
for _, row in df_nurses.iterrows():
    ano_inicio = int(row['Ano'])
    anos_exec = int(row['anos_execucao'])
    valor_total = row['precoContratual']
    
    if pd.notna(ano_inicio) and pd.notna(anos_exec) and anos_exec > 0:
        valor_anual = valor_total / anos_exec
        for i in range(anos_exec):
            ano = ano_inicio + i
            if ano in data_nursing:
                data_nursing[ano] += valor_anual
```

# Max and min values of data

```python
def feature_range(series: pd.Series):
    """
    Calculate min, max, and difference of a feature
    
    Parameters:
        series (pd.Series): The feature/column to analyze.
    
    Returns:
        dict: Dictionary with min, max, and difference.
    """
    min_val = series.min()
    max_val = series.max()
    diff = max_val - min_val
    
    return {
        "min": min_val,
        "max": max_val,
        "difference": diff
    }
```

```python
result = feature_range(df_nursing_byyear["Value"])
display(result)
```

```python
result = feature_range(df_medicin_byyear["Value"])
display(result)
```

```python
result = feature_range(df_publicadmin["Nurses"])
display(result)
```

```python
result = feature_range(df_publicadmin["Doctors"])
display(result)
```

# Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_outliers(ax, df, feature, title, ylabel, palette="Blues"):
    """
    Plot a boxplot highlighting potential outliers.
    """
    sns.boxplot(
        data=df,
        y=feature,
        color=palette[0],
        ax=ax,
        flierprops={"marker":"o", "markerfacecolor":"red", "markersize":6}  # highlight outliers
    )
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlabel("")

# ---- Create 2x2 grid ----
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

plot_outliers(axs[0,0], df_nursing_byyear, "Value", "Nursing Contracts (Value)", "Value", colors)
plot_outliers(axs[0,1], df_medicin_byyear, "Value", "Medical Contracts (Value)", "Value", colors)
plot_outliers(axs[1,0], df_publicadmin, "Nurses", "Nurses in Public Administration", "Nurses", colors)
plot_outliers(axs[1,1], df_publicadmin, "Doctors", "Doctors in Public Administration", "Doctors", colors)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def identify_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

df_nursing_byyear["is_outlier"] = identify_outliers(df_nursing_byyear["Value"])

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df_nursing_byyear,
    x="Year",
    y="Value",
    hue="is_outlier",
    palette={False:"blue", True:"red"}
)
plt.title("Nursing Contracts (Value) with Outliers Highlighted")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```

# Comparison - Dual Y-axis plot (original values)

Pros:
- preserves original magnitudes
- easy to see absolute values and trends at the same time
- no normalization required

Cons:
- can be confusing if scales differ too mutch
- readers may misinterpret trends if y-axes are not clearly labeled

Best use case:
- comparing datasets with different units or scales, when both absolute values matter

```python
# Merge keeping original values
df_compare_nurses = df_nursing_byyear.merge(
    df_publicadmin[["year", "Nurses"]],
    left_on="Year", right_on="year"
)

df_compare_doctors = df_medicin_byyear.merge(
    df_publicadmin[["year", "Doctors"]],
    left_on="Year", right_on="year"
)

# Nurses
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.plot(df_compare_nurses["Year"], df_compare_nurses["Value"], marker="o", color=colors[0], label="Nursing Contracts")
ax1.set_xlabel("Year")
ax1.set_ylabel("Nursing Contracts", color=colors[0])
ax1.tick_params(axis='y', labelcolor=colors[0])

ax2 = ax1.twinx()
ax2.plot(df_compare_nurses["Year"], df_compare_nurses["Nurses"], marker="s", color=colors[3], label="Nurses in Public Admin")
ax2.set_ylabel("Nurses in Public Admin", color=colors[3])
ax2.tick_params(axis='y', labelcolor=colors[3])

plt.title("Nurses Contracts vs Public Admin Nurses (Original Scale)")
fig.tight_layout()
plt.show()

# Doctors
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.plot(df_compare_doctors["Year"], df_compare_doctors["Value"], marker="o", color=colors[0], label="Medical Contracts")
ax1.set_xlabel("Year")
ax1.set_ylabel("Medical Contracts", color=colors[0])
ax1.tick_params(axis='y', labelcolor=colors[0])

ax2 = ax1.twinx()
ax2.plot(df_compare_doctors["Year"], df_compare_doctors["Doctors"], marker="s", color=colors[3], label="Doctors in Public Admin")
ax2.set_ylabel("Doctors in Public Admin", color=colors[3])
ax2.tick_params(axis='y', labelcolor=colors[3])

plt.title("Medical Contracts vs Public Admin Doctors (Original Scale)")
fig.tight_layout()
plt.show()
```

# Comparison - Normalize by first-year value (baseline year)

Pros:
- highlights growth/decline relative to a starting point
- makes trends directly comparable regardless of magnitude
- simple interpretation: "how much has it grown?"

Cons:
- loses information about absolute values
- sensitive to an unusual first-year value (outliers)

Best use case:
- comparing relative trends over the time or pecentage change from a reference

```python
# Nurses
df_compare_nurses["Value_base"] = df_compare_nurses["Value"] / df_compare_nurses["Value"].iloc[0]
df_compare_nurses["Nurses_base"] = df_compare_nurses["Nurses"] / df_compare_nurses["Nurses"].iloc[0]

plt.figure(figsize=(10,6))
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Value_base"], marker="o",color=colors[0], label="Nursing Contracts (Baseline 1st Year)")
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Nurses_base"], marker="s",color=colors[3], label="Public Admin Nurses (Baseline 1st Year)")
plt.title("Nurses Contracts vs Public Admin Nurses (Relative to First Year)")
plt.xlabel("Year")
plt.ylabel("Relative to First Year")
plt.xticks(df_compare_nurses["Year"])
plt.legend()
plt.grid(True)
plt.show()

# Doctors
df_compare_doctors["Value_base"] = df_compare_doctors["Value"] / df_compare_doctors["Value"].iloc[0]
df_compare_doctors["Doctors_base"] = df_compare_doctors["Doctors"] / df_compare_doctors["Doctors"].iloc[0]

plt.figure(figsize=(10,6))
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Value_base"], marker="o",color=colors[0], label="Medical Contracts (Baseline 1st Year)")
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Doctors_base"], marker="s",color=colors[3], label="Public Admin Doctors (Baseline 1st Year)")
plt.title("Medical Contracts vs Public Admin Doctors (Relative to First Year)")
plt.xlabel("Year")
plt.ylabel("Relative to First Year")
plt.xticks(df_compare_doctors["Year"])
plt.legend()
plt.grid(True)
plt.show()
```

# Comparison - Logarithmic scale

Pros:
- compresses large differences, making trends in small and large values visible together
- good for exponential growth or orders-of-magnitude differences

Cons:
- values below zero cannot be plotted
- less intuitive for casual readers; slope interpretation differs

Best use case:
- when data differ greatly in magnitude and you want to see proportional trends

```python
# Nurses
plt.figure(figsize=(10,6))
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Value"], marker="o",color=colors[0], label="Nursing Contracts")
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Nurses"], marker="s",color=colors[3], label="Public Admin Nurses")
plt.yscale("log")
plt.title("Nurses Contracts vs Public Admin Nurses (Log Scale)")
plt.xlabel("Year")
plt.ylabel("Value (Log Scale)")
plt.xticks(df_compare_nurses["Year"])
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

# Doctors
plt.figure(figsize=(10,6))
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Value"], marker="o",color=colors[0], label="Medical Contracts")
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Doctors"], marker="s",color=colors[3], label="Public Admin Doctors")
plt.yscale("log")
plt.title("Medical Contracts vs Public Admin Doctors (Log Scale)")
plt.xlabel("Year")
plt.ylabel("Value (Log Scale)")
plt.xticks(df_compare_doctors["Year"])
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

# Comparison - Min-max Normalization

Pros:
- All series are scaled to the same range, making visual comparison easy
- good for trend comparison

Cons:
- absolute magnitudes are lost
- sensitive to outliers (max value dominates scalling)

Best use case:
- comparing shape of trends ignoring absolute numbers

```python
df_compare_nurses["Value_norm"] = df_compare_nurses["Value"] / df_compare_nurses["Value"].max()
df_compare_nurses["Nurses_norm"] = df_compare_nurses["Nurses"] / df_compare_nurses["Nurses"].max()

df_compare_doctors["Value_norm"] = df_compare_doctors["Value"] / df_compare_doctors["Value"].max()
df_compare_doctors["Doctors_norm"] = df_compare_doctors["Doctors"] / df_compare_doctors["Doctors"].max()

# Nurses
plt.figure(figsize=(10,6))
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Value_norm"], marker="o",color=colors[0], label="Nursing Contracts (Normalized)")
plt.plot(df_compare_nurses["Year"], df_compare_nurses["Nurses_norm"], marker="s",color=colors[3], label="Public Admin Nurses (Normalized)")
plt.title("Comparison of Nurses Contracts vs Public Admin Nurses (Normalized)")
plt.xlabel("Year")
plt.ylabel("Normalized Value")
plt.xticks(df_compare_nurses["Year"])
plt.legend()
plt.grid(True)
plt.show()

# Doctors
plt.figure(figsize=(10,6))
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Value_norm"], marker="o",color=colors[0], label="Medical Contracts (Normalized)")
plt.plot(df_compare_doctors["Year"], df_compare_doctors["Doctors_norm"], marker="s",color=colors[3], label="Public Admin Doctors (Normalized)")
plt.title("Comparison of Medical Contracts vs Public Admin Doctors (Normalized)")
plt.xlabel("Year")
plt.ylabel("Normalized Value")
plt.xticks(df_compare_doctors["Year"])
plt.legend()
plt.grid(True)
plt.show()
```
